## Now let us plot some graphs to better analyze the given data

Let us analyse the scores of students in math,reading and writing.

#### Math Score

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['math score'],palette='Set1')
plt.title('Math Score',fontsize=22)
plt.show()
```

### Reading score

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['reading score'],palette='Set2')
plt.title('Reading Score',fontsize=22)
plt.show()
```

#### Writing Score

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['writing score'],palette='dark')
plt.title('Writing Score',fontsize=22)
plt.show()
```

```python
plt.subplot(141)
plt.title('Reading Score')
sns.violinplot(y='reading score',data=df,color='r',linewidth=1.5)
plt.subplot(142)
plt.title('Math Score')
sns.violinplot(y='math score',data=df,color='b',linewidth=1.5)
plt.subplot(143)
plt.title('Writing Score')
sns.violinplot(y='writing score',data=df,color='g',linewidth=1.5)
```

In the above given violin plots we can conclude that the maximum number of students have scored marks in the range of 60 - 80 in all of the three subjects.

```python
plt.subplot(131)
plt.title('Reading score')
sns.barplot(x='test preparation course',y='reading score',data=df)
plt.subplot(132)
plt.title('Math score')
sns.barplot(x='test preparation course',y='math score',data=df)
plt.subplot(133)
plt.title('Writing score')
sns.barplot(x='test preparation course',y='writing score',data=df)
plt.show()
```

We can see from the above bar graphs that those students who completed test preparation course have scored higher in all the three subjects than the students who did not.

```python
plt.title('Gender vs parental level of education',fontsize=20)
sns.countplot(x='gender',hue='parental level of education',data=df)
plt.show()
```

```python
plt.title('Race/ethnicity vs Parental level of education',fontsize=20)
sns.countplot(x='race/ethnicity',hue='parental level of education',data=df)
plt.show()
```

```python
plt.subplot(131)
plt.title('Reading score')
sns.barplot(x='gender',y='reading score',data=df)
plt.subplot(132)
plt.title('Math score')
sns.barplot(x='gender',y='math score',data=df)
plt.subplot(133)
plt.title('Writing score')
sns.barplot(x='gender',y='writing score',data=df)
plt.show()
```

From the above given bar graphs we can conclude that male students scored more in maths as compared to female students whereas female students scored more in reading and writing than male students.

```python
plt.figure(figsize=(20,20))
plt.subplots_adjust(left=0.125,bottom=0.1,right=1,top=1,wspace=0.6,hspace=0.3)



plt.subplot(141)
plt.title('Ethnicity',fontsize=18)
df['race/ethnicity'].value_counts().plot.pie(autopct="%1.1f%%")
plt.subplot(142)
plt.title('Parental level of education',fontsize=18)
df['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
plt.subplot(143)
plt.title('Gender',fontsize=18)
df['gender'].value_counts().plot.pie(autopct="%1.1f%%")
plt.subplot(144)
plt.title('Lunch',fontsize=18)
df['lunch'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()


```

* From the first pie chart of Ethnicity,we can see that majority of the students i.e 31.9% belong from group C
* From the third pie chart of gender we can conclude that number of female students in our dataset are slightly greater in number than female students
* From the fourth pie chart we can conclude that most of the students i.e 64.5% are provided with standard lunch.

```python


sns.set(style="whitegrid")
plt.subplot(131)
plt.title('Reading score')
sns.barplot(y='parental level of education',x='reading score',hue='parental level of education',data=df)
sns.set(style="whitegrid")
plt.subplot(132)
plt.title('Math score')
sns.barplot(y='parental level of education',x='math score',hue='parental level of education',data=df)
sns.set(style="whitegrid")
plt.subplot(133)
plt.title('Writing score')
sns.barplot(y='parental level of education',x='writing score',hue='parental level of education',data=df)



plt.show()
```

From the above graphs we can observe that students whose parental level of education is master's degree are scoring more marks in all the three subjects i.e maths,reading and writing whereas those students whose parental level of education is only high school degree have comparitively less score in all the three subjects.


```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=0.125,bottom=0.01,right=1.5,top=1,wspace=0.5,hspace=0.2)


plt.subplot(251)
plt.title('Test Preparation course vs Gender',fontsize=15)
sns.countplot(hue='test preparation course',x='gender',data=df)
plt.subplot(252)
plt.title('Test Preparation course vs Lunch',fontsize=15)
sns.countplot(hue='test preparation course',x='lunch',data=df)
plt.subplot(253)
plt.title('Test Preparation course vs Parental level of eduaction',fontsize=15)
sns.countplot(hue='test preparation course',y='parental level of education',data=df)
plt.subplot(254)
plt.title('Test Preparation course vs Race/Ethnicity',fontsize=15)
sns.countplot(hue='test preparation course',y='race/ethnicity',data=df)
plt.show()

```

* From the test preparation vs gender graph we can conclude that majority of both male as well as female students have not completed test preparation course.
* Standard lunch students have completed test preparation course
* Students whose parental level of education is 'some college','high school' and 'associate's degree' have completed test preparation course
* A large number of students who belong to group C have completed their test preparation course.

To analyse the data in a better way, we should add some new columns such as percentage and total score

```python
df['total score']=df['math score']+df['reading score']+df['writing score']
```

```python
df['percentage']=df['total score']/100*100
```

```python
plt.subplot(131)
plt.title('Total score vs parental level of education')
sns.barplot(y='parental level of education',x='total score',hue='parental level of education',data=df)
plt.subplot(132)
plt.title('Total score vs Gender')
sns.barplot(y='total score',x='gender',hue='gender',data=df)
plt.subplot(133)
plt.title('Total score vs Lunch')
sns.barplot(y='total score',x='lunch',hue='lunch',data=df)
plt.show()

```

* From total score vs parental level of education bar chart,we can conclude that students with parents habing master's degree and bachelor's degree have scored comparitively greater overall marks.
* From total score vs Gender bar chart,we can see that female students have scored higher overall marks than male students.
* From total score vs lunch bar chart,we can see that Students having standard lunch have higher overall score.

```python
plt.subplot(131)
plt.title('Total score vs Race/ethnicity')
sns.barplot(y='race/ethnicity',x='total score',hue='race/ethnicity',data=df)
plt.subplot(132)
plt.title('Total score vs test preparation corse')
sns.barplot(y='total score',x='test preparation course',hue='test preparation course',data=df)
plt.show()
```

* From the Total score vs Race/ethnicity bar chart , we can conclude that Group E students have scored higher overall score as compared to other groups.
* From total score vs Test preparation bar chart,students who have completed test preparation course have scored higher marks.
